Filtering songs according to a list of keywords

Data: https://data.mendeley.com/datasets/3t9vbwxgr5

```python
import pandas as pd
```

```python
df = pd.read_csv('Music Dataset Lyrics and Metadata from 1950 to 2019/tcc_ceds_music.csv', index_col=0)
df.head
```

```python
keywords = ['ice', 'snow', 'boat', 'sea', 'air', 'tree', 'power', 'energy']
```

```python
#filtered_df = df[df['track_name'].str.contains('|'.join(keywords))]
pattern = r'\b(?:{})\b'.format('|'.join(keywords))

# Filter the DataFrame
filtered_df = df[df['track_name'].str.contains(pattern, regex=True)]
filtered_df
```

```python
filtered_df.to_csv('minichallenge_songs.csv')
```
